```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
import pandas as pd


from nltk.tokenize import sent_tokenize, word_tokenize
y= pd.read_csv('ttt2_0.csv')
text = y.loc[:,'Примечание']
```

```python
ss=text.fillna("xxx")
```

```python
ss.head(15)
```

```python
a=[]
for i in ss:
    a.append(word_tokenize(i, language="russian"))
a
```

```python
from nltk.corpus import stopwords
```

```python
stopwords = nltk.corpus.stopwords.words('russian')
stopwords.append('вх')
stopwords.append('об')
stopwords.append('-')
stopwords.append('"')
stopwords.append('``')
stopwords.append(',')
stopwords.append('.')
stopwords.append('»')
stopwords.append('«')
stopwords.append('(')
stopwords.append('\'\'')
stopwords.append(')')
```

```python
new=a
from nltk.stem import SnowballStemmer
snowball = SnowballStemmer(language="russian")
for x in range(0,len(new)):
    for y in range(0,len(new[x])):
        ss = new[x][y]
        new[x][y]=snowball.stem(ss)
print(new)
```

```python
stop_words = stopwords
filtered_str=[]
new2=[]
for x in range(0,len(new)):
    for y in range(0,len(new[x])):
        if new[x][y]=='xxx':
            new[x][y]=' '
        filtered_str=' '.join((filter(lambda s: s not in stop_words,new[x])))
    new2.append(filtered_str)
    print(filtered_str)
```

```python
se = pd.Series(new2)
se
```

```python
p=pd.DataFrame({'Примечание':new2})
p.head(50)
```

```python
ee=y.drop(['Примечание'], axis = 1)
```

```python
rrr=ee.join(p)
```

```python
rrr.to_csv('ttt2_1.csv', index=False)
```
