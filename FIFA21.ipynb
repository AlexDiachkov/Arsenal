```python
import pandas as pd
df= pd.read_csv('fifa21_male2.csv')
pd.options.display.max_columns = 1000
df.head(5)
```

```python
df=df.sort_values("OVA")
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import median
import matplotlib.pyplot as plt
```

```python
columns=['Name', 'Age', 'Nationality','Club','OVA','POT','BP','Height',
         'Weight','foot','Value','Wage','Release Clause','IR']
```

```python
df = df[columns]
```

```python
df
```

```python
df["potencial"] = 0
df
```

```python
for i in range(0,len(df["potencial"])):
    df.potencial[i]=df.POT[i]-df.OVA[i]
```

```python
df
```

```python
df_year = df.copy()
plt.figure(figsize=(20,7))
sns.countplot(data= df_year,x="Age")
plt.show()
```

```python
plt.figure(figsize=(30,10))
plt.xticks(rotation= -35)
sns.barplot(data= df_year,x="Age", y= "BP",estimator=median, palette='hls')
plt.show()
```

```python
df_year['Nationality'].value_counts().head(17).plot.bar(figsize=(27,8),rot=0, title="Top 17 models")
plt.show()
```

```python
plt.figure(figsize=(15,4))
sns.countplot(data=df_year,x="BP")
plt.show()
```

```python
plt.figure(figsize=(15,4))
sns.countplot(data=df_year, x="type", hue="fuel")
plt.show()
```
