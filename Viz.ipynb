```python
import pandas as pd
import matplotlib.pyplot as plt
data = [['E001', 'M', 34, 123, 'Normal', 350],

        ['E002', 'F', 40, 114, 'Overweight', 450],

        ['E003', 'F', 37, 135, 'Obesity', 169],

        ['E004', 'M', 30, 139, 'Underweight', 189],

        ['E005', 'F', 44, 117, 'Underweight', 183],

        ['E006', 'M', 36, 121, 'Normal', 80],

        ['E007', 'M', 32, 133, 'Obesity', 166],

        ['E008', 'F', 26, 140, 'Normal', 120],

        ['E009', 'M', 32, 133, 'Normal', 75],

        ['E010', 'M', 36, 133, 'Underweight', 40] ]

df = pd.DataFrame(data, columns = ['EMPID', 'Gender', 

                                    'Age', 'Sales',

                                    'BMI', 'Income'] )
```

```python
df.plot.bar()

plt.bar(df['Age'], df['Sales'])

plt.xlabel("Age")

plt.ylabel("Sales")

plt.show()
```

```python
df.plot.box()

plt.boxplot(df['Income'])

plt.show()
```

```python
fig=plt.figure()
ax = fig.add_subplot(1,1,1)

ax.hist(df['Age'],bins = 7)
plt.title('Age distribution')
plt.xlabel('Age')
plt.ylabel('#Employee')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
fig=plt.figure()
ax = fig.add_subplot(1,1,1)
#Variable
ax.boxplot(df['Age'])
plt.show()
```

```python
import seaborn as sns 
sns.violinplot(df['Gender'], df['Age'])
sns.despine()
```

```python
var = df.groupby('Gender').Sales.sum() #grouped sum of sales at Gender level
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.set_xlabel('Gender')
ax1.set_ylabel('Sum of Sales')
ax1.set_title("Gender wise Sum of Sales")
var.plot(kind='bar')
```

```python
var = df.groupby('BMI').Sales.sum()
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.set_xlabel('BMI')
ax1.set_ylabel('Sum of Sales')
ax1.set_title("BMI wise Sum of Sales")
var.plot(kind='line')
```

```python
var = df.groupby(['BMI','Gender']).Sales.sum()
var.unstack().plot(kind='bar',stacked=True,  color=['red','blue'], grid=False)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(df['Age'],df['Sales'])
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(df['Age'],df['Sales'], s=df['Income']) # Added third variable income as size of the bubble
plt.show()
```

```python
var=df.groupby(['Gender']).sum().stack()
temp=var.unstack()
type(temp)
x_list = temp['Sales']
label_list = temp.index
plt.pie(x_list,labels=label_list,autopct="%1.1f%%") 
plt.title("Pastafarianism expenses") 
plt.show()
```

```python
import numpy as np
data = np.random.rand(4,2)
rows = list('1234') 
columns = list('MF') 
fig,ax=plt.subplots()
ax.pcolor(data,cmap=plt.cm.Reds,edgecolors='k')
ax.set_xticks(np.arange(0,2)+0.5)
ax.set_yticks(np.arange(0,4)+0.5)
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()
ax.set_xticklabels(columns,minor=False,fontsize=20)
ax.set_yticklabels(rows,minor=False,fontsize=20)
plt.show()
```
