```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
import pandas as pd

a=[]
from nltk.tokenize import sent_tokenize, word_tokenize
y= pd.read_csv('ttt.csv', delimiter=';', encoding="cp1251")
text = y.loc[:,'Краткое\nсодержание']
for i in text:
    a.append(word_tokenize(i, language="russian"))
a
```

```python
from nltk.corpus import stopwords
```

```python
stopwords = nltk.corpus.stopwords.words('russian')
stopwords.append('вх')
stopwords.append('об')
stopwords.append('-')
stopwords.append('"')
stopwords.append('``')
stopwords.append(',')
stopwords.append('.')
stopwords.append('»')
stopwords.append('«')
stopwords.append('(')
stopwords.append('\'\'')
stopwords.append(')')
stopwords
```

```python
new=a
from nltk.stem import SnowballStemmer
snowball = SnowballStemmer(language="russian")
for x in range(0,len(new)):
    for y in range(0,len(new[x])):
        ss = new[x][y]
        new[x][y]=snowball.stem(ss)
```

```python
stop_words = stopwords
filtered_str=[]
new2=[]
for x in range(0,len(new)):
    for y in range(0,len(new[x])):
        filtered_str=' '.join((filter(lambda s: s not in stop_words,new[x])))
        #filtered_str=filtered_str.split()
        #if (y<6):
        #    new2.append(filtered_str)
        #print(filtered_str)
    new2.append(filtered_str)
    
```
